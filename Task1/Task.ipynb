```python
import pandas as pd
import numpy as np
```

```python
def read_return_top10(filename):
    return pd.read_csv(filename)

df = read_return_top10('Military Expenditure.csv')
df.head(10)
```

```python
df.shape
```

```python
# Ջնջել այն տողերը որոնց 'Type' սյունը "Country" չէ։
```

```python
def remove_non_countries(df):
    df.drop(df[df['Type'] != 'Country'].index, inplace=True)

remove_non_countries(df)
df.head(10)
```

```python
# Ջնջել 'Code', 'Type', 'Indicator Name' սյուները։
```

```python
def remove_some_cols(df):
    df.drop(columns=['Code', 'Type', 'Indicator Name'], inplace=True)

remove_some_cols(df)
df.head(10)
```

```python
# Ստացված dataframe-ը գրել(պահպանել) 'MilitaryExpenditureSmall.csv' ֆայլի վրա այնպես, ինչպես որ կա։
```

```python
def save_df(df, filename):
    df.to_csv(filename, index=False)

save_df(df, 'MilitaryExpenditureSmall.csv')
```

```python
# Վերադարձնել Հայաստանին համապատասխանող տողի այն տվյալները/վանդակները որոնք դատարկ չեն։
```

```python
def info_arm(df):
    armenia_row = df[df['Name'] == 'Armenia']
    return armenia_row.iloc[0].dropna()

info = info_arm(df)
info
```

```python
# Վերադարձնել DataFrame, որտեղ կլինի պետության անունը՝ որպես ինդեքս և յուրաքանչյուր պետության միջին տարեկան ծախսը՝ որպես 'mean_expense'
# սյուն՝ գրված նվազման կարգով։ Վերջնական աղյուսակում չպետք է լինեն այնպիսի երկրներ, որոնց ծախսերի տվյալները ընդհանրապես առկա չեն կամ 0 են։
```

```python
def per_country_mean_expense(df):
    data_columns = df.iloc[:, 1:]
    row_means = data_columns.mean(axis=1)
    result_df = pd.DataFrame({'Country': df['Name'], 'mean_expense': row_means})
    result_df = result_df[result_df['mean_expense'].notna() & (result_df['mean_expense'] != 0)]
    result_df = result_df.sort_values(by='mean_expense', ascending=False)
    result_df = result_df.set_index('Country')
    return result_df

res = per_country_mean_expense(df)
res
```
